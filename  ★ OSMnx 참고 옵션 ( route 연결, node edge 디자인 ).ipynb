```python
import networkx as nx
import osmnx as ox
import folium
```

## **1. Gaph import & Option**

```python
G = ox.graph_from_place('해운대구, 부산, 대한민국', network_type='drive') #network type : ‘walk’, ‘bike’, ‘drive’, ‘drive_service’, ‘all’
fig, ax = ox.plot_graph(G, bgcolor="k",  #배경색
                        edge_color='green',   #엣지 컬러
                        node_color ='white',   #노드 컬러
                        node_size=8,     #노드 사이즈
                        edge_linewidth=1,   #엣지 굵기
                        edge_alpha=1,     # 
                        figsize=(20,20), dpi=300, save=True)   #픽셀 사이즈 ,해상도, 저장여부 
```

## **2. Routes - OSMnx**

```python
orig_node = ox.nearest_nodes(G,129.12976213893924,35.173065429000914) #에이스하이테크21
dest_node = ox.nearest_nodes(G,129.16362922754982,35.16307508780792) #해운대구청

route = nx.shortest_path(G, orig_node, dest_node, weight='length')
```

```python
orig_node1= ox.nearest_nodes(G,129.1198649235281, 35.18836037060419, return_dist=False)  #재송역
dest_node1= ox.nearest_nodes(G,129.1241107600455,35.179418666196916, return_dist=False )  #센텀역

orig_node2 = ox.nearest_nodes(G,129.1241107600455,35.179418666196916, return_dist=False)  #센텀역
dest_node2 = ox.nearest_nodes(G,129.13731436680837,35.1706073957403, return_dist=False ) #벡스코역

orig_node3 = ox.nearest_nodes(G,129.13731436680837,35.1706073957403, return_dist=False) #벡스코역
dest_node3 = ox.nearest_nodes(G,129.1587325652332,35.16381130295812, return_dist=False ) #해운대역

route1 = nx.shortest_path(G,orig_node1,dest_node1, weight='length')
route2 = nx.shortest_path(G,orig_node2,dest_node2, weight='length')
route3 = nx.shortest_path(G,orig_node3,dest_node3, weight='length')
routes = [route1,route2,route3 ]
```

```python
#최단거리 추출

len_km = nx.shortest_path_length(G, orig_node1, dest_node1, weight='length') / 1000    #km단위로 도출
print(round(len_km,4), "km")
```

```python
fig, ax = ox.plot_graph_routes(G, routes,route_colors=['g','r','y'],    # route 색을 리스트 형태로, 경로 별로 설정할 수 있음
                               node_size=3, edge_linewidth=2,
                               route_linewidth=10,
                               figsize=(15,15), dpi=300)
```
