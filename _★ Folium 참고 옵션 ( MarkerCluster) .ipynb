```python
import folium 
```

```python
#베이스맵 띄우기
test_map = folium.Map(location = [35.16376167626317, 129.15291524631846],
                      zoom_start=12.5,tiles ="CartoDB positron")
#test_map
```

```python
import pandas as pd

sk_data = pd.read_csv('해운대, 백스코 호텔 20230311.csv')
sk_data.head(10)
```

```python
from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(test_map)

for i in range(len(sk_data)):
    sk_data_Marker = folium.Marker([sk_data.iloc[i,8],sk_data.iloc[i,7]],).add_to(marker_cluster)
    
test_map
```
