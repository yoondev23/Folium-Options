## **0. RESET**

```python
#기본 zoom 위치 설정
test_map = folium.Map(location = [35.1576059507137, 129.05917617141483],
                      zoom_start=12.5)
test_map
```

## **1. Marker Style Option**

```python
#option1 기본 marker

marker011 = folium.Marker( [35.1576059507137, 129.05917617141483],   
                      popup = '마커이름', #마커 클릭시 팝업
                      icon = folium.Icon(color='blue'))

marker012 = folium.Marker( [35.15815680415907, 129.07547407354517],   
                      tooltip = '마우스오버',  #마우스오버 표시
                      icon = folium.Icon(color='blue'))

marker013 = folium.Marker( [35.15837318026466, 129.09013956458466 ],   
                      popup = '마커이름',
                      tooltip = '마우스오버',  
                      icon = folium.Icon(color='blue'))   #중복해서 사용가능

marker011.add_to(test_map)
marker012.add_to(test_map)
marker013.add_to(test_map)
test_map
```

```python
#option2 원형 marker

marker02 = folium.CircleMarker( [35.1576059507137, 129.05917617141483],
                       radius = 50,
                       color = '#147AA6',
                       popup = '마커이름',
                       fill_color = 'skyblue')

marker02.add_to(test_map)
test_map
```

## **2. Icon style option**

```python
#option1 기본 marker

home = folium.Marker( [35.14852164454257, 129.05921052934232],   
                      icon=folium.Icon(icon = 'home', color = 'purple',icon_color ='black'))

star = folium.Marker( [35.14863043556692, 129.06974618119276],   
                      icon=folium.Icon(icon = 'star', color = 'red'))

flag = folium.Marker( [35.1487983125548, 129.08098553473704],   
                      icon=folium.Icon(icon = 'flag', color = 'green'))

bookmark = folium.Marker( [35.14904617650333, 129.09197643504356],   
                      icon=folium.Icon(icon = 'bookmark', color = 'gray'))

info_sign = folium.Marker( [35.14939431713456, 129.1007499938054],   
                      icon=folium.Icon(icon = 'info-sign', color = 'pink'))

home.add_to(test_map)
star.add_to(test_map)
flag.add_to(test_map)
bookmark.add_to(test_map)
info_sign.add_to(test_map)
test_map
```
